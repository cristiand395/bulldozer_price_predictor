```python
# Lets parse 'salesdate' reimporting the dataset
df = pd.read_csv('./data/TrainAndValid.csv',
                 low_memory=False,
                 parse_dates=["saledate"])
df.saledate
```

Now its the date column its parsed

```python
# Sorting by Date
df.sort_values(by = ['saledate'], 
               inplace = True, 
               ascending = True)
df.head()
```

```python
# Create a copy of the dataset
df_copy = df.copy()
```

```python
# Split SaleDate by year, month, day..
df_copy['saleYear'] = df_copy.saledate.dt.year
df_copy['saleMonth'] = df_copy.saledate.dt.month
df_copy['saleDay'] = df_copy.saledate.dt.day
df_copy['saleDayOfWeek'] = df_copy.saledate.dt.dayofweek
df_copy['saleDayOfYear'] = df_copy.saledate.dt.dayofyear
```

```python
# drop saledate column because now its splited
df_copy.drop('saledate', axis=1, inplace=True)
```

```python
df_copy.head().T
```

```python
df_copy.info()
```

Need to change Object columns to strings

```python
# Using Pandas manipulation, lest change object columns into string columns
for label, content in df_copy.items():
  if pd.api.types.is_string_dtype(content):
    df_copy[label] = content.astype('category').cat.as_ordered()
```

### Fill Missing Values

```python
df_copy.isna().sum()
```

```python
# Check which columns have numeric data and missing values and filing with the median
for label, content in df_copy.items():
  if pd.api.types.is_numeric_dtype(content):
    if pd.isnull(content).sum():
      df_copy[label] = content.fillna(content.median())
```

```python
# Check which columns dont have numeric data and missing values and filing with the median
for label, content in df_copy.items():
  if not pd.api.types.is_numeric_dtype(content):
    df_copy[label] = pd.Categorical(content).codes+1
```

```python
df_copy.isna().sum()
```

Great no missing values!!
Now because there are too many data, we are going to split it in train and validate set to work better.

```python
# Split data into validation and training data
df_val = df_copy[df_copy.saleYear == 2012]
df_train = df_copy[df_copy.saleYear != 2012]
```

```python
# Split data into X and y sets
X_train, y_train = df_train.drop("SalePrice", axis=1), df_train.SalePrice
X_valid, y_valid = df_val.drop("SalePrice", axis=1), df_val.SalePrice
```

```python
  # Create evaluation function (the competition uses RMSLE)
from sklearn.metrics import mean_squared_log_error, mean_absolute_error, r2_score

def rmsle(y_test, y_preds):
    """
    Caculates root mean squared log error between predictions and
    true labels.
    """
    return np.sqrt(mean_squared_log_error(y_test, y_preds))

# Create function to evaluate model on a few different levels
def show_scores(model):
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_valid)
    scores = {"Training MAE": mean_absolute_error(y_train, train_preds),
              "Valid MAE": mean_absolute_error(y_valid, val_preds),
              "Training RMSLE": rmsle(y_train, train_preds),
              "Valid RMSLE": rmsle(y_valid, val_preds),
              "Training R^2": r2_score(y_train, train_preds),
              "Valid R^2": r2_score(y_valid, val_preds)}
    return scores
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instantiate the model and set max samples because of the size of the dataset
model = RandomForestRegressor(n_jobs=-1,
                              random_state=42,
                              max_samples=10000)
# n_jobs set to -1 to use all the processors of my pc
```

```python
model.fit(X_train, y_train)
```
